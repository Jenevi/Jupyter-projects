```python
import torch
import math
import random

import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
class TriangleCircle4x4Dataset(torch.utils.data.Dataset):
    def __init__(self, size=1000):
        self.size = size
    # говорит каким может быть максимальный index для __getitem__
    def __len__(self):
        return self.size

    # возвращает элемент датасета под номером index
    def __getitem__(self, index):
        a = torch.tensor([[0,1],[1,1]])
        b = torch.tensor([[1,0],[1,1]])
        c = torch.tensor([[1,1],[1,0]])
        d = torch.tensor([[1,1],[0,1]])

        aa = torch.tensor([[1,0],[0,0]])
        bb = torch.tensor([[0,1],[0,0]])
        cc = torch.tensor([[0,0],[0,1]])
        dd = torch.tensor([[0,0],[1,0]])

        if random.randint(0, 1):
            return (random.choice([a,b,c,d]).float(), 1)
        else:
            return (random.choice([aa,bb,cc,dd]).float(), 0)
```

```python
dataset = TriangleCircle4x4Dataset(1000)
data_loader = torch.utils.data.DataLoader(
    # drop_last=True для отбрасывания того, что не поместилось в батч
    # num_workers=0 для однопроцессорной загрузки
    dataset, batch_size=4, shuffle=True, num_workers=0, drop_last=True)

# for i, batch in enumerate(data_loader):
#     print(i, batch)
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.inp = nn.Flatten()
        self.fc1 = nn.Linear(4, 10)
        self.fc2 = nn.Linear(10, 10)
        self.fc3 = nn.Linear(10, 10)
        self.fc4 = nn.Linear(10, 2)

    def forward(self, x):
#         print(x.shape)
        x = self.inp(x)
#         print(x.shape)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = F.relu(self.fc4(x))
# #         return F.log_softmax(x)
#         return x
        return F.softmax(x, dim=1)
```

```python
network = Net()
learning_rate = 1e-3
momentum = 0.9
optimizer = optim.SGD(network.parameters(), lr=learning_rate,
                      momentum=momentum)
```

```python
learning_rate = 1e-3
momentum = 0.9
optimizer = optim.SGD(network.parameters(), lr=learning_rate,
                      momentum=momentum)
```

```python
# def train(epoch):
network.train()
criterion = nn.CrossEntropyLoss()
for i in range(100):
    for batch_idx, (data, target) in enumerate(data_loader):
        optimizer.zero_grad()
        output = network(data)
        loss = criterion(output, target)
        print('loss: ', loss)
        loss.backward()
        optimizer.step()
```
